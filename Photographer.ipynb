```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
PG=pd.read_csv("Photographer.csv",encoding='cp1252')
```

```python
PG.head()
```

```python
PG = PG.drop("S.no",axis=1)
```

```python
PG.head()
```

```python
nan_count = PG.isna().sum()

print("Number of NaN values:", nan_count)
```

```python
Group = PG.groupby("State")
```

```python
Group
```

```python
Group.head()
```

```python
Group.first()
```

```python
na_counts = PG.groupby('State')['Website'].apply(lambda x: x.isna().sum())

print(na_counts)
```

```python
state_counts = PG['State'].value_counts()

print(state_counts)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# Bar chart for NaN counts
axes[0].bar(na_counts.index, na_counts.values)
axes[0].set_title('NaN Counts Statewise')
axes[0].set_xlabel('State')
axes[0].set_ylabel('Count')

# Bar chart for state counts
axes[1].bar(state_counts.index, state_counts.values)
axes[1].set_title('State Counts')
axes[1].set_xlabel('State')
axes[1].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# Bar chart for NaN counts
axes[0].bar(na_counts.index, na_counts.values)
axes[0].set_title('NaN Counts Statewise')
axes[0].set_xlabel('State')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(na_counts.index, rotation='vertical')
# Bar chart for state counts
axes[1].bar(state_counts.index, state_counts.values)
axes[1].set_title('State Counts')
axes[1].set_xlabel('State')
axes[1].set_ylabel('Count')
axes[1].set_xticklabels(state_counts.index, rotation='vertical')
plt.tight_layout()
plt.show()
```

```python
nan_counts = PG['Phone no.'].isna().groupby(PG['State']).sum()

print(nan_counts)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# Bar chart for NaN counts
axes[0].bar(nan_counts.index, nan_counts.values)
axes[0].set_title('NaN Counts Statewise')
axes[0].set_xlabel('State')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(na_counts.index, rotation='vertical')
# Bar chart for state counts
axes[1].bar(state_counts.index, state_counts.values)
axes[1].set_title('State Counts')
axes[1].set_xlabel('State')
axes[1].set_ylabel('Count')
axes[1].set_xticklabels(state_counts.index, rotation='vertical')
plt.tight_layout()
plt.show()
```
